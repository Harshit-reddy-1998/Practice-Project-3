```python
for col in df_new.columns:
    percentiles = df_new[col].quantile([0.01, 0.99]).values
    df_new[col][df_new[col] <= percentiles[0]] = percentiles[0]
    df_new[col][df_new[col] >= percentiles[1]] = percentiles[1]
```

```python
#outlier analysis
for column_name in df_new.columns:
    q1 = df_new[column_name].quantile(0.25)
    q3 = df_new[column_name].quantile(0.75)
    iqr = q3-q1
    upper = q3 +1.5*iqr
    lower = q1 -1.5*iqr
    outliner_df = df_new.loc[(df_new[column_name] < lower)| (df_new[column_name] > upper)]
    print('Percentage of outlier in {0} is {1}'.format(column_name, outliner_df.shape[0]/df_new.shape[0]))
```

```python
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
maxAcc=0
maxRS=0
for i in range(1,200):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.35,random_state=i)
    mod=DecisionTreeClassifier()
    mod.fit(x_train,y_train)
    pred=mod.predict(x_test)
    acc=accuracy_score(y_test,pred)
    if acc>maxAcc:
        maxAcc=acc,
        maxRS=i
print('Best accuracy is',maxAcc,'on random state',maxRS)        
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.35,random_state=145)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
dtc=DecisionTreeClassifier()
dtc.fit(x_train,y_train)
pred=dtc.predict(x_test)
print(accuracy_score(y_test,pred))
print(confusion_matrix(y_test,pred))
print(classification_report(y_test,pred))
```

```python
mnb=MultinomialNB()
mnb.fit(x_train,y_train)
pred=mnb.predict(x_test)
print(accuracy_score(y_test,pred))
print(confusion_matrix(y_test,pred))
print(classification_report(y_test,pred))
```

```python
knn=KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train,y_train)
pred=knn.predict(x_test)
print(accuracy_score(y_test,pred))
print(confusion_matrix(y_test,pred))
print(classification_report(y_test,pred))
```

```python
svc=SVC()  # By default kernel=rbf
svc.fit(x_train,y_train)
pred=svc.predict(x_test)
print(accuracy_score(y_test,pred))
print(confusion_matrix(y_test,pred))
print(classification_report(y_test,pred))
```

These, accuracy scores can be becaues of overfitting. Lets, check with cross-validation.

# Cross-Validation

```python
from sklearn.model_selection import cross_val_score
score=cross_val_score(mnb,x,y,cv=5)
print(score)
print(score.mean())
```

```python
score=cross_val_score(dtc,x,y,cv=5)
print(score)
print(score.mean())
```

```python
score=cross_val_score(knn,x,y,cv=5)
print(score)
print(score.mean())
```

```python
score=cross_val_score(svc,x,y,cv=5)
print(score)
print(score.mean())
```

The difference between the accuracy and cv says that svc is working better compared to the other models.

# Hyper parameter tuning

We use this technique to improve the accuracy of the best model that we have choosen.

```python
from sklearn.model_selection import GridSearchCV
parameters={'kernel':['linear','rbf','poly','sigmoid'],
            'C':[1,5,10]}
```

```python
gs=GridSearchCV(SVC(),parameters,cv=5)
gs.fit(x_train,y_train)
```

```python
print(gs.best_params_)
```

```python
Final_model=SVC(kernel='linear')
Final_model.fit(x_train,y_train)
pred=Final_model.predict(x_test)
acc=accuracy_score(y_test,pred)
print(acc*100)
```
